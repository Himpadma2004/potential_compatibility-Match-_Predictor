```python
def simplify_categories(df):
    # ---- Diet ----
    diet_mapping = {
        'mostly vegetarian': 'vegetarian',
        'strictly vegetarian': 'vegetarian',
        'mostly vegan': 'vegan',
        'strictly vegan': 'vegan',
        'mostly anything': 'anything',
        'strictly anything': 'anything',
        'mostly halal': 'halal',
        'strictly halal': 'halal',
        'mostly kosher': 'kosher',
        'strictly kosher': 'kosher',
        'mostly other': 'other',
        'strictly other': 'other'
    }
    df['diet'] = df['diet'].replace(diet_mapping)

    # ---- Drinks ----
    df['drinks'] = df['drinks'].replace({
        'not at all': 'never',
        'rarely': 'occasionally',
        'socially': 'occasionally',
        'very often': 'often',
        'desperately': 'often'
    })

    # ---- Drugs ----
    df['drugs'] = df['drugs'].replace({
        'sometimes': 'occasionally',
        'often': 'frequently'
    })

    # ---- Smokes ----
    df['smokes'] = df['smokes'].replace({
        'when drinking': 'sometimes',
        'trying to quit': 'sometimes'
    })

    # ---- Religion ----
    religion_mapping = {
        'catholicism': 'christianity'
    }
    df['religion'] = df['religion'].replace(religion_mapping)
    df['religion'] = df['religion'].str.extract(r'(\w+)', expand=False)

    # ---- Education ----
    education_mapping = {
        'high school': 'highschool',
        'working on high school': 'highschool',
        'graduated from high school': 'highschool',
        'dropped out of high school': 'highschool',
        'two-year college': 'college',
        'working on two-year college': 'college',
        'graduated from two-year college': 'college',
        'dropped out of two-year college': 'college',
        'college/university': 'college',
        'working on college/university': 'college',
        'graduated from college/university': 'college',
        'dropped out of college/university': 'college',
        'masters program': 'masters',
        'working on masters program': 'masters',
        'graduated from masters program': 'masters',
        'dropped out of masters program': 'masters',
        'ph.d program': 'phd',
        'working on ph.d program': 'phd',
        'graduated from ph.d program': 'phd',
        'dropped out of ph.d program': 'phd',
        'law school': 'law',
        'working on law school': 'law',
        'graduated from law school': 'law',
        'dropped out of law school': 'law',
        'med school': 'med',
        'working on med school': 'med',
        'graduated from med school': 'med',
        'dropped out of med school': 'med',
        'space camp': 'other',
        'working on space camp': 'other',
        'graduated from space camp': 'other',
        'dropped out of space camp': 'other'
    }
    df['education'] = df['education'].replace(education_mapping)

    # ---- Job ----
    job_mapping = {
        'banking / financial / real estate': 'finance',
        'computer / hardware / software': 'tech',
        'sales / marketing / biz dev': 'sales/marketing',
        'medicine / health': 'healthcare',
        'science / tech / engineering': 'STEM',
        'education / academia': 'education',
        'clerical / administrative': 'admin',
        'construction / craftsmanship': 'construction',
        'political / government': 'government',
        'law / legal services': 'legal',
        'retired': 'other',
        'transportation': 'transport',
        'hospitality / travel': 'hospitality',
        'artistic / musical / writer': 'creative',
        'entertainment / media': 'media',
        'executive / management': 'management',
        'military': 'military',
        'rather not say': 'unknown'
    }
    df['job'] = df['job'].replace(job_mapping)

    # ---- Pets ----
    pets_mapping = {
        'likes dogs and likes cats': 'likes dogs and cats',
        'has dogs and likes cats': 'has dogs and cats',
        'likes dogs and has cats': 'has dogs and cats',
        'has dogs and has cats': 'has dogs and cats',
        'has cats': 'has cats',
        'has dogs': 'has dogs',
        'dislikes dogs and dislikes cats': 'dislikes dogs and cats',
        'dislikes dogs and has cats': 'dislikes dogs and cats',
        'likes dogs and dislikes cats': 'likes dogs and dislikes cats',
        'dislikes dogs and likes cats': 'dislikes dogs and likes cats',
    }
    df['pets'] = df['pets'].replace(pets_mapping)

    return df

df_cleaned = simplify_categories(df)
```

```python
import pandas as pd
import re
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
import numpy as np

# Download NLTK resources (safe download)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

def clean_text(text):
    """
    Cleans and preprocesses text data from the essay columns.
    """
    if not isinstance(text, str):
        return ""
    
    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Tokenize
    words = text.split()
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    
    return " ".join(words)

# --- Demo dataset ---
data = {
    'essay0': ["about me: i am an intellectual who loves to talk about ideas.", "i'm a chef and I work in a restaurant."],
    'essay1': ["currently working as a developer. trying to make cool stuff.", np.nan],
}
essays_df = pd.DataFrame(data)

# Combine essay columns
essay_cols = [col for col in essays_df.columns if col.startswith("essay")]
essays_df['all_essays'] = essays_df[essay_cols].fillna('').apply(lambda row: ' '.join(row.astype(str)), axis=1)

# Clean
essays_df['all_essays_cleaned'] = essays_df['all_essays'].apply(clean_text)

print(essays_df[['all_essays', 'all_essays_cleaned']])
```

```python
def prepare_essay_text(df, essay_cols=None):
    """Combine essay columns into single text column 'essay'."""
    if essay_cols is None:
        # common names: essay0..essay9
        essay_cols = [c for c in df.columns if c.startswith("essay")]
    df = df.copy()
    # join essays, handle NaNs
    df['essay'] = df[essay_cols].fillna('').agg(' '.join, axis=1).str.strip()
    return df
```

```python
def norm(x, minv=None, maxv=None):
    """Normalize scalar x into 0-1 given min and max; if not given, infer from x."""
    if minv is None or maxv is None:
        return x  # caller handles
    if maxv == minv:
        return 0.0
    return (x - minv) / (maxv - minv)
```

```python
def shared_hobbies_score(h1, h2):
    # If hobbies are text lists separated by ';' or comma, adapt accordingly
    set1 = set()
    set2 = set()
    if isinstance(h1, str) and h1:
        set1 = set([s.strip().lower() for s in h1.replace(',', ';').split(';') if s.strip()])
    if isinstance(h2, str) and h2:
        set2 = set([s.strip().lower() for s in h2.replace(',', ';').split(';') if s.strip()])
    if not set1 and not set2:
        return 0.0
    inter = len(set1 & set2)
    union = len(set1 | set2) if len(set1 | set2) > 0 else 1
    return inter / union  # Jaccard-style similarity 0-1
```

```python
def categorical_match_score(v1, v2):
    """Return 1.0 if exact match (and not 'unknown'), else 0."""
    if pd.isna(v1) or pd.isna(v2): return 0.0
    if str(v1).strip().lower() in ('', 'unknown', 'nan') or str(v2).strip().lower() in ('', 'unknown', 'nan'):
        return 0.0
    return 1.0 if str(v1).strip().lower() == str(v2).strip().lower() else 0.0
```

```python
def personality_similarity(pvec1, pvec2):
    # pvec = [O,C,E,A,N] floats 0-1 ideally. Compute normalized distance -> similarity.
    a = np.array(pvec1, dtype=float)
    b = np.array(pvec2, dtype=float)
    if np.isnan(a).any() or np.isnan(b).any():
        # fallback: if missing, return 0.5 neutral
        return 0.5
    dist = np.linalg.norm(a - b)  # range up to sqrt(5) for 0-1 scale
    max_dist = np.sqrt(5)
    sim = max(0.0, 1 - (dist / max_dist))
    return float(sim)
```

```python
def essay_similarity(essay1, essay2, tfidf_vec=None):
    # Accepts pre-fit tfidf vectorizer or will compute on the two essays
    if tfidf_vec is None:
        vec = TfidfVectorizer(max_features=2000, stop_words='english')
        X = vec.fit_transform([essay1 or "", essay2 or ""])
        cos = cosine_similarity(X[0:1], X[1:2])[0][0]
        return float(np.nan_to_num(cos, 0.0)), vec
    else:
        X = tfidf_vec.transform([essay1 or "", essay2 or ""])
        cos = cosine_similarity(X[0:1], X[1:2])[0][0]
        return float(np.nan_to_num(cos, 0.0)), None
```

```python
def conflict_penalty(profileA, profileB):
    """
    Penalize obvious conflicts (binary) that should reduce compatibility:
    - One wants children (offspring) but other doesn't â€” example
    - Strong opposite on religion that cause conflict (optional)
    Returns penalty in [0,1] to subtract from final score.
    Customize based on available columns.
    """
    penalty = 0.0
    # Example: offspring field contains 'wants kids' / 'doesn't want' etc. Adjust to your column style
    for col in ['offspring']:
        if col in profileA.index and col in profileB.index:
            a = str(profileA[col]).lower() if pd.notna(profileA[col]) else ''
            b = str(profileB[col]).lower() if pd.notna(profileB[col]) else ''
            # if one says 'wants kids' and other 'doesn't want' or 'doesn't have'
            if ('wants' in a and 'not' in b) or ('wants' in b and 'not' in a):
                penalty += 0.3
    # clamp
    return min(0.9, penalty)
```

```python
def compatibility_score(profileA, profileB, tfidf_vec=None,
                        weights=None, essay_weight_method='similarity'):
    """
    Compute compatibility in [0,100].
    weights: dict with keys 'sim', 'person', 'comp', 'essay', 'conflict_penalty'
    essay_weight_method: 'similarity' or 'complement' (use essay similarity as similarity or complement)
    """
    if weights is None:
        weights = {'sim':0.5, 'person':0.3, 'comp':0.2, 'essay':0.25, 'conflict_penalty':1.0}
        # Note: essay weight is applied inside sim/component calculation so final sum should be normalized later.
    # Ensure expected fields exist by using .get or default values
    # 1) Similarity components (categorical exact matches + hobbies + movie/music)
    cat_cols = ['sex','orientation','diet','drinks','drugs','education','pets','religion','smokes']
    cat_scores = []
    for c in cat_cols:
        if c in profileA.index and c in profileB.index:
            cat_scores.append(categorical_match_score(profileA[c], profileB[c]))
    cat_sim = np.mean(cat_scores) if cat_scores else 0.0

    # hobbies (if exist)
    h_score = 0.0
    if 'hobbies' in profileA.index and 'hobbies' in profileB.index:
        h_score = shared_hobbies_score(profileA['hobbies'], profileB['hobbies'])

    # movie/music matches if exist
    movie_sim = 0.0
    music_sim = 0.0
    if 'fav_movie_genre' in profileA.index and 'fav_movie_genre' in profileB.index:
        movie_sim = categorical_match_score(profileA['fav_movie_genre'], profileB['fav_movie_genre'])
    if 'fav_music_genre' in profileA.index and 'fav_music_genre' in profileB.index:
        music_sim = categorical_match_score(profileA['fav_music_genre'], profileB['fav_music_genre'])

    # combine categorical and hobby/movie/music into S_sim:
    # give hobbies and movie/music slightly higher weight inside sim
    S_sim = np.mean([cat_sim, h_score, movie_sim, music_sim])

    # 2) Personality similarity
    pcols = ['O','C','E','A','N']
    if all(c in profileA.index for c in pcols) and all(c in profileB.index for c in pcols):
        pvecA = [profileA[c] for c in pcols]
        pvecB = [profileB[c] for c in pcols]
        S_person = personality_similarity(pvecA, pvecB)
    else:
        S_person = 0.5  # neutral if unknown

    # 3) Essay similarity (semantic) â€” normalized 0-1
    essayA = profileA.get('essay', '') if 'essay' in profileA.index else ''
    essayB = profileB.get('essay', '') if 'essay' in profileB.index else ''
    essay_sim, fitted_vec = essay_similarity(essayA, essayB, tfidf_vec=tfidf_vec)

    # 4) Complementarity measure: simple measure based on differences that aren't conflicts.
    # We define comp = 1 - (average categorical match), so if people are different, comp is higher.
    # But we should damp comp where differences are likely negative (we rely on conflict_penalty)
    S_comp = 1.0 - S_sim  # simple; you can refine per feature

    # Combine: prefer similarity and personality, but allow some complement
    w_sim = weights.get('sim', 0.5)
    w_person = weights.get('person', 0.3)
    w_comp = weights.get('comp', 0.2)
    # incorporate essay similarity into similarity bucket
    # combine S_sim and essay_sim into combined_sim
    combined_sim = (S_sim * (1 - weights.get('essay', 0.25))) + (essay_sim * weights.get('essay', 0.25))

    raw = (w_sim * combined_sim) + (w_person * S_person) + (w_comp * S_comp)
    # apply conflict penalty
    penalty = conflict_penalty(profileA, profileB) * weights.get('conflict_penalty', 1.0)
    raw = raw - penalty
    raw = np.clip(raw, 0.0, 1.0)
    return raw * 100.0, fitted_vec  # returns percent and fitted tfidf if created

############################################################
# Example usage on DataFrame `df`:
# 1) ensure essays combined
# 2) pick two profile indices or ids and compute
############################################################
```

```python
def compute_pair_compatibility(df, idxA, idxB, essay_cols=None, weights=None, reuse_tfidf=None):
    df2 = prepare_essay_text(df, essay_cols=essay_cols)
    profileA = df2.loc[idxA]
    profileB = df2.loc[idxB]
    score, fitted_vec = compatibility_score(profileA, profileB, tfidf_vec=reuse_tfidf, weights=weights)
    # if fitted_vec returned and reuse_tfidf is None, you can re-run with the fitted vector for speed on many pairs
    return score, fitted_vec
```

```python
# If you want to compute pair compatibility across the dataset for many pairs,
# fit a TF-IDF on all essays once:
def fit_tfidf_on_all_essays(df, essay_cols=None):
    df2 = prepare_essay_text(df, essay_cols=essay_cols)
    vec = TfidfVectorizer(max_features=5000, stop_words='english')
    vec.fit(df2['essay'].fillna('').astype(str).tolist())
    return vec
```

```python
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Step 0: ensure df_selected exists

# Step 1: Encode categorical columns
categorical_cols = ["sex","orientation","diet","drinks","drugs","education","job","pets","religion","smokes"]
encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df_selected[col] = df_selected[col].astype(str)
    df_selected[col] = le.fit_transform(df_selected[col])
    encoders[col] = le

# Step 2: Fit TF-IDF on text column
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
tfidf_matrix = tfidf_vectorizer.fit_transform(df_selected['essay'])

# Step 3: Fit K-Means
kmeans_model = KMeans(n_clusters=5, random_state=42)
kmeans_model.fit(tfidf_matrix)

# Step 4: Save everything
model_package = {
    "encoders": encoders,
    "tfidf": tfidf_vectorizer,
    "kmeans": kmeans_model,
    "columns": df_selected.columns
}

with open("final_model.pkl", "wb") as f:
    pickle.dump(model_package, f)

print("âœ… final_model.pkl saved successfully!")
```
